```python
from scipy import stats

def find_outliers_z(df):
    outliers_dict = {}
    for column in df.select_dtypes(include=[float, int]).columns:
        z_scores = stats.zscore(df[column])
        outliers = (z_scores > 3) | (z_scores < -3)
        if outliers.any():
            outliers_dict[column] = df[outliers]
    return outliers_dict
```

```python
outliers_z = find_outliers_z(df)
print("Kolone sa autlajerima koristeći Z-Score metodu:")
for column, data in outliers_z.items():
    print(f"Kolona: {column}, Broj autlajera: {len(data)}")
```

```python
def find_outliers_iqr(df):
    outliers_dict = {}
    for column in df.select_dtypes(include=[float, int]).columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        outliers = (df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR))
        if outliers.any():
            outliers_dict[column] = df[outliers]
    return outliers_dict
```

```python
outliers_iqr = find_outliers_iqr(df)
print("Kolone sa autlajerima koristeći IQR metodu:")
for column, data in outliers_iqr.items():
    print(f"Kolona: {column}, Broj autlajera: {len(data)}")
```
